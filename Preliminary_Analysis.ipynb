```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web

#Show our plots
%matplotlib inline
#ggplot
style.use('ggplot')
```

```python
'''
theorize that stocks which have small % inside ownership, 
along with small % institutional ownership, are more 
responsive to changes in public sentiment (bc they are
relatively entirely owned by the public). These stocks
also have high volatility, and are priced at >$1.
''';
```

```python
#We're going to watch a ~ 2 year time frame.

start = dt.datetime(2016,1,1)
end = dt.datetime(2018,12,31)
```

## Picking a couple of stocks, w/ above criteria.
### Later, add more controls for things like size, volume, industry.

```python
ESDA_Data = web.DataReader('EDSA','yahoo',start,end)
GWGH_Data = web.DataReader('GWGH','yahoo',start,end)
PT_Data = web.DataReader('PT','yahoo',start,end)
OXBR_Data = web.DataReader('OXBR','yahoo',start,end)
IMUX_Data = web.DataReader('IMUX','yahoo',start,end)
SPEX_Data = web.DataReader('SPEX','yahoo',start,end)
```

```python
list_of_stocks = [ESDA_Data,GWGH_Data,PT_Data,
                  OXBR_Data,IMUX_Data,SPEX_Data]
```

```python
'''look at everything together'''

def plot_everything(list_of_dataframe):
    for element in list_of_dataframe:
        plt.plot(element.Close)
    
plot_everything(list_of_stocks)
```
